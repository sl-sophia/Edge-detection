```python
#edge detector
import cv2
import numpy as np
import os
import glob
input_files=glob.glob("/mnt/C2A15796657DA486/db/validation/labels/*")
for f in range(len(input_files)):
    file=input_files[f]
    name=os.path.basename(os.path.splitext(file)[0])
    img = cv2.imread(file,cv2.IMREAD_GRAYSCALE)
    _img=np.zeros((len(img),len(img[0])))
    for i in range(len(img)-1):
        for j in range(len(img[0])-1):
            if img[i][j]!=img[i+1][j]:
                _img[i][j]=255
                _img[i+1][j]=255
                if(i+2<len(img)):
                    _img[i+2][j]=255
            if img[i][j+1]!=img[i][j]:
                _img[i][j]=255
                _img[i][j+1]=255
                if(j+2<len(img[0])):
                    _img[i][j+2]=255
    cv2.imwrite("/mnt/CF07D760A1B7847F/edges/"+name+".png",_img)
```

```python
#edge detector
#This enables choosing the thickness
import cv2
import numpy as np
import os
import glob
input_files=glob.glob("/mnt/CF07D760A1B7847F/edges/test/*")
for f in range(len(input_files)):
    file=input_files[f]
    name=os.path.basename(os.path.splitext(file)[0])
    file="/mnt/C2A15796657DA486/db/validation/labels/"+name+".png"
    img = cv2.imread(file,cv2.IMREAD_GRAYSCALE)
    width=len(img[0])
    height=len(img)
    _img=np.zeros((len(img),len(img[0])))
    thick=3
    for i in range(len(img)-1):
        for j in range(len(img[0])-1):
            if img[i][j]!=img[i+1][j] and (img[i][j]==84 or img[i+1][j]==84 or img[i][j]==76 or img[i+1][j]==76):
                x=i+1
                y=j
                for k in range(max(0,x-thick),min(height,x+thick)):
                    for l in range(max(0,y-thick),min(width,y+thick)):
                        _img[k][l]=255
            if img[i][j+1]!=img[i][j] and (img[i][j]==84 or img[i][j+1]==84 or img[i][j]==76 or img[i+1][j]==76):
                x=i
                y=j+1
                for k in range(max(0,x-thick),min(height,x+thick)):
                    for l in range(max(0,y-thick),min(width,y+thick)):
                        _img[k][l]=255
    cv2.imwrite("/mnt/CF07D760A1B7847F/eee/"+name+".png",_img)
```

```python
#convert detected gray to black&white
import cv2
import numpy as np
import os
import glob
input_files=glob.glob("/mnt/CF07D760A1B7847F/sophia/test/*0.png")
for f in range(len(input_files)):
    file=input_files[f] 
    img = cv2.imread(file,cv2.IMREAD_GRAYSCALE)
    name=os.path.basename(os.path.splitext(file)[0])
    for i in range(len(img)):
        for j in range(len(img[0])):
            if img[i][j]>50:
                img[i][j]=0
            else:
                img[i][j]=255
    cv2.imwrite("/mnt/CF07D760A1B7847F/output/"+name+".png",img)
```

```python
#Accuracy
input_files=glob.glob("/mnt/CF07D760A1B7847F/output/*")
se=np.zeros(400)
sp=np.zeros(400)
setotal=0
sptotal=0
count_contains_white=0
for f in range(len(input_files)):
    file=input_files[f] 
    img = cv2.imread(file,cv2.IMREAD_GRAYSCALE)
    name=os.path.basename(os.path.splitext(file)[0])
    label=cv2.imread("/mnt/CF07D760A1B7847F/ed_resize/"+name+".png",cv2.IMREAD_GRAYSCALE)
    white_common=0
    white_label=0
    contains_white=0
    for i in range(len(img)):
        for j in range(len(img[0])):
            if img[i][j]>50 and label[i][j]>50:
                white_common+=1
            if label[i][j]>50:
                white_label+=1
                contains_white=1
    if contains_white==0:
        continue
    else:
        count_contains_white+=1
    se[f]=white_common/white_label
    sp[f]=(153600-white_common)/(153600-white_label)
    print("se:",se[f],"sp[f]",sp[f])
    setotal+=se[f]
    sptotal+=sp[f]
print("----------------------------")
print("setotal",setotal,"sptotal",sptotal)
sp_avg=sptotal/count_contains_white
se_avg=setotal/count_contains_white
print("se_avg",se_avg,"sp_avg",sp_avg)
```
